```python
df1['numbercarsowned '] = zscore(df1['NumberCarsOwned'])
df1['NumberChildrenathome'] = zscore(df1['NumberChildrenAtHome'])
df1['totalChildren'] = zscore(df1['TotalChildren'])
df1['yearlyincome '] = zscore(df1['YearlyIncome'])
```

```python
X = ['NumberCarsOwned','TotalChildren','YearlyIncome','NumberChildrenAtHome']
df2 = df1.drop(X,axis = 1) 
```

```python
df2.head()
```

```python
df2['Gender'] = df2['Gender'].apply(lambda x : 0 if x == 'M' else 1)
df2['MaritalStatus'] = df2['MaritalStatus'].apply(lambda x : 0 if x == 'M' else 1)
```

```python
df2.head()
```

```python
df2.Education.unique()
```

```python
def edu(x):
    if x=='Bachelors':
        x= 3
    elif x == 'Graduate Degree':
        x = 3
    elif x=='Partial College':
        x= 2 
    elif x=='High School':
        x= 1 
    else:
        x= 0
    return x
```

```python
df2['Education'] = df2['Education'].apply(edu)
```

```python
df2.head()
```

```python
dummies = pd.get_dummies(df2.Occupation)
dummies1 = pd.get_dummies(df2.CountryRegionName)
dummies2 = pd.get_dummies(df2.StateProvinceName)
```
